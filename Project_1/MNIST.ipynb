## Comparing activation functions

```python
network1 = MLP([500], sigmoid, init='Xavier', bias_presence=True, eta=0.01,
              alpha=0.9, max_epochs=150, regression=False, random_state=1, epochs_no_change=20)
```

```python
network1.fit(x, y, evaluation_dataset=[x_test, y_test], calculate_accuracy=True)
print("Accuracy on training data: " + str(accuracy(network1.predict(x), y)))
print("Accuracy on test data: " + str(accuracy(network1.predict(x_test), y_test)))
plot_errors_vs_epochs(network1.accuracy, None, 'accuracy')
```

```python
network2 = MLP([500], tanh, init='Xavier', bias_presence=True, eta=0.01,
              alpha=0.9, max_epochs=150, regression=False, random_state=1, epochs_no_change=20)
```

```python
network2.fit(x, y, evaluation_dataset=[x_test, y_test], calculate_accuracy=True)
print("Accuracy on training data: " + str(accuracy(network2.predict(x), y)))
print("Accuracy on test data: " + str(accuracy(network2.predict(x_test), y_test)))
plot_errors_vs_epochs(network2.accuracy, None, 'accuracy')
```

```python
network3 = MLP([500], relu, init='Xavier', bias_presence=True, eta=0.01,
              alpha=0.9, max_epochs=150, regression=False, random_state=1, epochs_no_change=20)
```

```python
network3.fit(x, y, evaluation_dataset=[x_test, y_test], calculate_accuracy=True)
print("Accuracy on training data: " + str(accuracy(network3.predict(x), y)))
print("Accuracy on test data: " + str(accuracy(network3.predict(x_test), y_test)))
plot_errors_vs_epochs(network3.accuracy, None, 'accuracy')
```

```python
def plot_acc(acc1, acc2, acc3):
    plt.figure()
    x = [x for x in range(max(len(acc1),len(acc2),len(acc3)))]
    plt.plot(x[:len(acc1)], acc1, 'red')
    plt.plot(x[:len(acc2)], acc2, 'blue')
    plt.plot(x[:len(acc3)], acc3, 'black')
    plt.title("Accuracy change during training for different activation functions")
    plt.legend(["Sigmoid","Hyperbolic tangent","ReLU"])
    plt.xlabel("Epochs")
    plt.ylabel("Accuracy")
    plt.show()
plot_acc(network1.accuracy, network2.accuracy, network3.accuracy)
```
